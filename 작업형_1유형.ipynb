###1. crim 항목의 상위에서 10번째 값 (상위 10개의 값 중에서 가장 적은 값)으로 상위 10개의 값을 변환하고, age가 80 이상인 값에 대하여 crim의 평균을 구하시오

```python
import numpy as np
import pandas as pd
```

```python
#  데이터 불러오기
df = pd.read_csv('/content/mydrive/MyDrive/Colab_Notebooks/빅분기/데이터 셋/P210201.csv')
```

```python
# 불필요한 변수 삭제
df = df.drop('Unnamed: 0', axis=1)
```

```python
# 내림차순으로 정렬하여 10개 항목에 대해 top10에 저장
top10 = df['crim'].sort_values(ascending=False).head(10)
print(top10)
```

```python
# top10의 10번째 값을 tenth 변수에 저장
tenth = top10.iloc[9]
```

```python
#  df의 crim 값이 tenth 변수보다 크거나 같으면 tenth 변수 값으로 변경
df['crim'] = np.where(df['crim']>=tenth, tenth, df['crim'])
```

```python
# age가 80 이상인 행만 추출하여 저장
over80 = df[df['age']>80]
```

```python
# 요약 통계량 출력
print(over80['age'].describe())
```

```python
# 문제에서 요구한 값 구하기
print(round(over80['crim'].mean(),2))
```

###3. charges 항목에서 이상값의 합을 구하시오(이상값은 평균에서 1.5 표준편차 이상인 값으로 하고 답안은 소수점 이하는 버리고 정수로 출력)

```python
df = pd.read_csv('/content/mydrive/MyDrive/Colab_Notebooks/빅분기/데이터 셋/P210203.csv')
df.head()
```

```python
# 평균 + 1.5*std
uo = np.mean(df['charges']) + 1.5 * np.std(df['charges'])

# 평균 - 1.5*std
lo = np.mean(df['charges']) - 1.5 * np.std(df['charges'])
```

```python
result = df[(df['charges']>=uo)| (df['charges']<=lo)]['charges'].sum()
print(int(result))
```

###13.  출동소방서별 신고일시로부터 출동일시까지의 연도별 월평균을 구하고, 가장 늦게 출동한 출동소방서의 월평균 시간을 분단위로 제출형식에 맞게 제출 (시간은 30초 단위로 반올림)

```python
# 데이터 불러오기
df = pd.read_csv('/content/mydrive/MyDrive/Colab_Notebooks/빅분기/데이터 셋/P230601.csv')
df.head()
```

```python
from datetime import datetime
```

```python
df['call_time'] = pd.to_datetime(df['신고일시'])
df['arr_time'] = pd.to_datetime(df['출동일시'])
df['fire_station'] = df['출동소방서']

df['diff_time'] = (df['arr_time'] - df['call_time']).dt.total_seconds()
```

```python
df = df.groupby([df['fire_station'], df['call_time'].dt.year, df['call_time'].dt.month]).mean('diff_time')
df = df.sort_values('diff_time', ascending = False)
```

```python
res_date = df['diff_time'].head(1)
res_num = float(res_date.iloc[0]) / 60
result = round(res_num)
print(result)
```
